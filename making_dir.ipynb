```
import os
import shutil
import glob
```

```
key = os.listdir("/content/drive/My Drive/ddsm_patches3")
num = 781
file = []
for i in range(len(key)):
  new_key = os.listdir("/content/drive/My Drive/ddsm_patches3/"+str(num))
  file.append(new_key)
  num+=1

cancer = "/content/drive/My Drive/patches_sort/cancer"
normal = "/content/drive/My Drive/patches_sort/normal"
benign = "/content/drive/My Drive/patches_sort/benign"
c,n,b = 0, 0, 0
flag = 781
for i in range(len(file)):
  for j in range(len(file[i])):
    if file[i][j].startswith("cancer"):
      file[i][j] = "/content/drive/My Drive/ddsm_patches3/"+str(flag)+"/"+file[i][j]
      shutil.copy(file[i][j], cancer)
      c+=1
    elif file[i][j].startswith("normal"):
      file[i][j] = "/content/drive/My Drive/ddsm_patches3/"+str(flag)+"/"+file[i][j]
      shutil.copy(file[i][j], normal)
      n+=1
    else:
      file[i][j] = "/content/drive/My Drive/ddsm_patches3/"+str(flag)+"/"+file[i][j]
      shutil.copy(file[i][j], benign)
      b+=1
    print("cancer = {}, normal = {}, benign = {}" .format(c, n, b))
  flag+=1
   
```
